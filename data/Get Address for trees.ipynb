```python
import pandas as pd
import numpy as np
import requests
```

```python
# function to get request from api
def getAddress(lat,lon):
    url = 'https://nominatim.openstreetmap.org/reverse'
    params = {
        "format":"json",
        "lat":lat,
        "lon":lon  
    }
    response = requests.get(url,params=params)
    return response.json()
```

```python
df = pd.read_csv('trees.csv')
```

```python
theSplits = np.array_split(df, 5)
hamza = theSplits[0]
mark = theSplits[1]
adneya = theSplits[2]
vikie = theSplits[3]
jaidv = theSplits[4]

hamza.head()
```

```python
theDict = dict(zip(hamza['CoM ID'],hamza['CoordinateLocation'].apply(lambda x: [float(y) for y in x[1:len(x)-1].split(",")] )))

fullAddress = {}
suburb = {}
for items in theDict:
    lat = theDict[items][0]
    lon = theDict[items][1]
    response = getAddress(lat,lon)
    fullAddress[items] = response['display_name']
    suburb[items] = response['address']['suburb']
```
